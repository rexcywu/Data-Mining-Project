```
##Load the train csv file

raw_data = pd.read_csv("/content/drive/My Drive/FTEC_4003_Project/Task_2/train.csv")
raw_test_data = pd.read_csv("/content/drive/My Drive/FTEC_4003_Project/Task_2/assignment-test.csv")
```

```
## Drop RowNumber, CustomerId, Surname
train_data = raw_data.drop(columns=['RowNumber', 'CustomerId', 'Surname'])
test_data = raw_test_data.drop(columns=['RowNumber', 'CustomerId', 'Surname'])
```

```
from sklearn.preprocessing import OneHotEncoder, LabelEncoder, StandardScaler
from scipy.sparse import csr_matrix
```

```
train_data['NumOfProducts'].unique()
```

```
for column in ['Geography', 'Gender']:
  train_data[column] = train_data[column].apply(str)

  #Label Encoder classify the categorical data into integer
  label_encoder = LabelEncoder()
  label = label_encoder.fit_transform(train_data[column])
  label = label.reshape(len(label), 1)

  #OneHotEncoder encode the integer generated from LabelEncoder
  one_hot_encoder = OneHotEncoder(sparse=False)
  encoded_col = one_hot_encoder.fit_transform(label)

  encoded_col = pd.DataFrame(encoded_col, columns = column + "_" + label_encoder.classes_)
  train_data = pd.concat([train_data, encoded_col], axis = 1)
  train_data = train_data.drop(columns = [column])
```

```
for column in ['Geography', 'Gender']:
  test_data[column] = test_data[column].apply(str)

  #Label Encoder classify the categorical data into integer
  label_encoder = LabelEncoder()
  label = label_encoder.fit_transform(test_data[column])
  label = label.reshape(len(label), 1)

  #OneHotEncoder encode the integer generated from LabelEncoder
  one_hot_encoder = OneHotEncoder(sparse=False)
  encoded_col = one_hot_encoder.fit_transform(label)

  encoded_col = pd.DataFrame(encoded_col, columns = column + "_" + label_encoder.classes_)
  test_data = pd.concat([test_data, encoded_col], axis = 1)
  test_data = test_data.drop(columns = [column])
```

```
test_data
```

```
scaler = StandardScaler()
standarized_features = ['CreditScore', 'Age','Tenure', 'Balance', 'NumOfProducts', 'EstimatedSalary']
train_data[standarized_features] = scaler.fit_transform(train_data[standarized_features])
test_data[standarized_features] = scaler.fit_transform(test_data[standarized_features])
```

```
train_data, test_data = train_data.align(test_data, join='outer', axis=1)
```

```
train_data.to_csv(r'/content/drive/My Drive/FTEC_4003_Project/Task_2/preprocessed_train_data.csv', index=False)
```
