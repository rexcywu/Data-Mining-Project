```
from scipy.sparse import csr_matrix
train_data = pd.read_csv("/content/drive/My Drive/FTEC_4003_Project/Task_2/preprocessed_train_data.csv")

test_data = pd.read_csv("/content/drive/My Drive/FTEC_4003_Project/Task_2/preprocessed_test_data.csv")

y_train = train_data['Exited']
X_train = train_data.drop(columns=['Exited'])
X_test = test_data.drop(columns=['Exited'])
X_test = X_test.fillna(0)
X_train = csr_matrix(X_train).todense()
X_test = csr_matrix(X_test).todense()
```

```
#Use SVM from SVC

from sklearn.svm import SVC
c_list = [0.00001, 0.0001, 0.001, 0.01, 0.1, 1, 10, 100]

from sklearn.metrics import classification_report, confusion_matrix, roc_curve
from sklearn import metrics
for chosen_c in c_list:
  print('c = ', chosen_c)
  clf = SVC(kernel='rbf', C = chosen_c)
  clf.fit(X_train,y_train)
  y_pred = clf.predict(X_train)
  print(confusion_matrix(y_train,y_pred))
  print(classification_report(y_train,y_pred))
  print(metrics.roc_auc_score(y_train,y_pred))
```

```
clf = SVC(kernel='rbf', C = 100)
clf.fit(X_train,y_train)
```

```
y_test_pred = clf.predict(X_test)
test_Response = pd.DataFrame(data=y_test_pred, columns=["Response"])
test_Response.index += 342999
test_Response.index.name = 'id'
test_Response.to_csv(r'/content/drive/My Drive/FTEC_4003_Project/Task_2/svc_test_result.csv', index=False)
```
